```python
import matplotlib.pyplot as plt
import torch

a = [1, -0.1, 0.5, -1, 0.2]
a = [1, -1, -1, -1, -1]
a_tensor = torch.tensor(a, dtype=torch.float32)
softmax_a = torch.nn.functional.softmax(a_tensor, dim=0).numpy()
temperature = 0.5
a_prime_tensor = a_tensor / temperature
softmax_a_prime = torch.nn.functional.softmax(a_prime_tensor, dim=0).numpy()
plt.bar(
    [i + 0.4 for i in range(len(a))],
    softmax_a_prime,
    width=0.4,
    label=f"Temperature={temperature}",
    color="orange",
)
plt.bar(range(len(a)), softmax_a, width=0.1, label="Temperature=1.0", color="blue")
plt.legend()
plt.bar(range(len(a)), softmax_a)
plt.xlabel("Index")
plt.ylabel("Softmax Value")
plt.title("Softmax of a")
plt.show()
```
